```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
from keras import Sequential
from keras.layers import Dense, Conv2D, MaxPooling2D, Flatten, Dropout, BatchNormalization
import os
from sklearn.model_selection import train_test_split
from keras.utils import to_categorical
```

```python
X_data_cats = os.listdir('cats/')
y_data_cats = [0 for i in range(len(X_data_cats))]
X_data_dogs = os.listdir('dogs/')
y_data_dogs = [1 for i in range(len(X_data_dogs))]
```

```python
df = pd.DataFrame(data=X_data_cats + X_data_dogs, columns=['Image'])
df['Target'] = y_data_cats + y_data_dogs
```

```python
df
```

```python
X = df.drop('Target', axis=1)
y = df['Target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15)
```

```python
mean_size_m = []
mean_size_n = []
for i in X_data_cats:
    image = mpimg.imread(f'cats/{i}')
    mean_size_m.append(image.shape[0])
    mean_size_n.append(image.shape[1])
for i in X_data_dogs:
    image = mpimg.imread(f'dogs/{i}')
    mean_size_m.append(image.shape[0])
    mean_size_n.append(image.shape[1])
```

```python
image_size_m = np.mean(mean_size_m)
image_size_n = np.mean(mean_size_n)
```

```python
image_size_m = int(np.round(image_size_m))
image_size_n = int(np.round(image_size_n))
```

```python
X_data_train = []
y_data_train = []
for i in range(0, len(X_train['Image'])):
    try:
        if y_train.iloc[i] == 0:
            image = mpimg.imread(f'cats/{X_train["Image"].iloc[i]}')
            image = cv2.resize(image, (image_size_m, image_size_n))
            image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
            image = image.reshape(image_size_m, image_size_n, 1)
            if X_train["Image"].iloc[i][-3:] == 'png':
                X_data_train.append(image)
                y_data_train.append(0)
            else:
                image = image / 255
                X_data_train.append(image)
                y_data_train.append(0)
                
        else:
            image = mpimg.imread(f'dogs/{X_train["Image"].iloc[i]}')
            image = cv2.resize(image, (image_size_m, image_size_n))
            image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
            image = image.reshape(image_size_m, image_size_n, 1)
            if X_train["Image"].iloc[i][-3:] == 'png':
                X_data_train.append(image)
                y_data_train.append(1)
            else:
                image = image / 255
                X_data_train.append(image)
                y_data_train.append(1)
    except:
        continue
            
X_data_train = np.array(X_data_train)
y_data_train = np.array(y_data_train)
```

```python
X_data_test = []
y_data_test = []
for i in range(0, len(X_test['Image'])):
    try:
        if y_test.iloc[i] == 0:
            image = mpimg.imread(f'cats/{X_test["Image"].iloc[i]}')
            image = cv2.resize(image, (image_size_m, image_size_n))
            image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
            image = image.reshape(image_size_m, image_size_n, 1)
            if X_test["Image"].iloc[i][-3:] == 'png':
                X_data_test.append(image)
                y_data_test.append(0)
            else:
                image = image / 255
                X_data_test.append(image)
                y_data_test.append(0)
                
        else:
            image = mpimg.imread(f'dogs/{X_test["Image"].iloc[i]}')
            image = cv2.resize(image, (image_size_m, image_size_n))
            image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
            image = image.reshape(image_size_m, image_size_n, 1)
            if X_test["Image"].iloc[i][-3:] == 'png':
                X_data_test.append(image)
                y_data_test.append(1)
            else:
                image = image / 255
                X_data_test.append(image)
                y_data_test.append(1)
    except:
        continue
            
X_data_test = np.array(X_data_test)
y_data_test = np.array(y_data_test)
```

```python
y_data_train = to_categorical(y_data_train)
y_data_test = to_categorical(y_data_test)
```

```python
model = Sequential()

model.add(Conv2D(filters=32, kernel_size=(3, 3), activation='relu', input_shape=(image_size_m, image_size_n, 1)))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(filters=64, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())

model.add(Dense(8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(2, activation='softmax'))


model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_data_train, y_data_train, validation_data=(X_data_test, y_data_test), epochs=5)
```
